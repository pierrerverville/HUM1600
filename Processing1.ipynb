```python
# Problem 1

# Write the code to prompt the user for two integers, x and y.

x = int(input("Enter an integer: "))
y = int(input("Enter a second integer: "))

# Then, write an if statement that compares the x and y and prints "X is larger than Y" only if that is true. If it is not true, do not print anything.

if x > y:
  print("X is larger than Y")
```

```python
# Problem 2
# Write the code to prompt the user for two integers, x and y.

x = int(input("Enter an integer: "))
y = int(input("Enter a second integer: "))

# Then, write an if-else statement that compares the x and y and prints "X is larger than Y" if that is true. Print "Y is larger than X" if it is not true.

if x > y:
  print("X is larger than Y")
else:
  print("Y is larger than X")

```

```python
# Problem 3

# Write the code to prompt the user for two integers, x and y.

x = int(input("Enter an integer: "))
y = int(input("Enter a second integer: "))

# Then, write an if-elif-else statement that compares the x and y and prints "X is larger than Y" if that is true, or prints "Y is larger than X" if that is true, or it prints "X and Y are equal" if that is true.

if x > y:
  print("X is larger than Y")
elif x < y:
  print("Y is larger than X")
else:
  print("X and Y are equal")
```

```python
# Problem 4

numbers = 1

while numbers <= 100:
    print(numbers)
    numbers += 1

# Write another while loop in the same Code cell that prints the integers from 100 down to 1.

numbers = 100

while numbers >= 1:
    print(numbers)
    numbers -= 1
```

```python
# Problem 5

#Write the code to print the numbers 1 to 100 using a for loop.

for numbers in range(1, 101):
  print(numbers)


#Write another for loop in the same Code cell that prints the integers from 100 down to 1.

for numbers in range(100, 0, -1):
  print(numbers)
```

```python
# Problem 6

# Write the code to print the numbers 10 through 100 using a for loop and the range() function.

for numbers in range(10, 101):
    print(numbers)
```
